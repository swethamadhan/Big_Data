```python
# 1. Automatic Answer Checker
# ------------------------
# !pip install nltk

# Code
import string
import nltk
similarityArray = []

# Splitting the string into individual tokens
def get_tokens(text):
  tokens = text.split()
  tokens = [token.lower() for token in tokens] # Lower casing all the tokens
  return tokens

# Similarity checking is done below using nltk's package ( jaccardian similarity range 0 to 1)
def calculate_similarity(answer, correct_answer):
  answer_set = set(get_tokens(answer))
  correct_answer_set = set(get_tokens(correct_answer))
  similarity = 1 - nltk.jaccard_distance(answer_set, correct_answer_set) # outputs the value of dissimilarity and hance we are subtracting 1
  return similarity

def check_answer(answer, correct_answer):
  similarity = calculate_similarity(answer, correct_answer)
  similarityArray.append(similarity)
  if similarity != 0:
    return True
  else:
    return False

Questions = {
    "1":"What is part of a database that holds only one type of information?",
    "2":"'OS' computer abbreviation usually means ?",
    "3":"What was the active medium used in the first working laser ever constructed?",
    "4":"'.MOV' extension refers usually to what kind of file?",
    "5":"What does EPROM stand for?"
}
CorrectAnswer = {
    "1":"Field",
    "2":"Operating System",
    "3":"A ruby rod",
    "4":"Animation movie file",
    "5":"Erasable Programmable Read Only Memory"
}
# User should provide their answers below in the values section of the UserAnswer dictionary

UserAnswer = {
    "1":"Field",
    "2":"Operation Subset",
    "3":"Titanium",
    "4":"Animation",
    "5":"Erasable Programmable Read Only Memory"
}
CountofQuestions = 5
for i in range(1,CountofQuestions+1):
  check_answer(UserAnswer.get(str(i)), CorrectAnswer.get(str(i)))

print(f"Your score is {sum(similarityArray)}")
```

```python
# 4. Online AI shopping assistant using M wallet system
# -----------------------------------------------------

# Code

# Define the available products and their prices
available_products = {
    "Smartphone": 500,
    "Headphones": 100,
    "Laptop": 800,
    "Camera": 300
}

# Initialize the wallet with an initial balance of 1000 credits
wallet_balance = 1000

# Initialize an empty shopping cart
shopping_cart = []

# Prompt the user to add products to the shopping cart
print("Welcome to the Online AI Shopping System!")
print("Available products:")
for product, price in available_products.items():
    print(f"{product}: {price} credits")

while True:
    user_input = input("Enter the product you want to buy (or 'q' to quit): ")

    if user_input.lower() == "q":
        break

    if user_input in available_products:
        shopping_cart.append(user_input)
        print(f"Added {user_input} to the shopping cart.")
    else:
        print("Invalid product. Please enter a valid product.")

# Calculate the total amount to be paid
total_amount = sum(available_products[product] for product in shopping_cart)

# Make a payment and update the wallet balance
if wallet_balance >= total_amount:
    wallet_balance -= total_amount
    print("Payment successful. Order placed!")
else:
    print("Insufficient balance. Payment failed.")

# Display the remaining wallet balance
print(f"Remaining wallet balance: {wallet_balance} credits")
```

```python
# 6.Online Plagiarism Checker
# ---------------------------

# Code

from difflib import SequenceMatcher

# Function to calculate similarity ratio between two texts
def calculate_similarity(text1, text2):
    similarity_ratio = SequenceMatcher(None, text1, text2).ratio()
    return similarity_ratio

# Chatbot loop
while True:
    user_input = input("Enter the assignment text to check for plagiarism (or 'q' to quit): ")

    if user_input.lower() == 'q':
        break

    try:
        # Prompt for comparison text
        comparison_text = input("Enter the text to compare with the assignment: ")

        # Calculate similarity
        similarity_ratio = calculate_similarity(user_input, comparison_text)
        print(f"Similarity ratio: {similarity_ratio}\n")
    except Exception as e:
        print(f"Error: {str(e)}")
```

```python
# 9. Teacher Automatic Time-Table Software
# ----------------------------------------

# Code

import random

# Define a list of teachers and their available subjects
teachers = {
    "Alice": ["Math", "Science"],
    "Bob": ["English", "History"],
    "Carol": ["Math", "Geography"],
    "David": ["Science", "Physics"],
}

# Define a list of days and available class timings
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
timings = ["9 AM", "10 AM", "11 AM", "2 PM", "3 PM"]

# Generate a random timetable based on teachers' preferences
def generate_timetable():
    timetable = {}
    for day in days:
        timetable[day] = {}
        for timing in timings:
            teacher = random.choice(list(teachers.keys()))
            subject = random.choice(teachers[teacher])
            timetable[day][timing] = {"teacher": teacher, "subject": subject}
    return timetable

# Print the generated timetable
def print_timetable(timetable):
    for day in days:
        print(f"\n{day}:")
        for timing in timings:
            class_info = timetable[day][timing]
            print(f"{timing} - {class_info['teacher']} - {class_info['subject']}")

# Run the automatic timetable generator
def main():
    print("Welcome to the Teacher's Automatic Timetable Software!")
    print("Generating Timetable...\n")

    timetable = generate_timetable()
    print_timetable(timetable)

if __name__ == "__main__":
    main()
```
